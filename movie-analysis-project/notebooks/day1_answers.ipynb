# Day 1 作业答案

```python
import pandas as pd
df = pd.read_csv('../data/movies.csv')
```

```python
# 作业 1: 找出评分最高的电影
highest_rated_idx = df['vote_average'].idxmax()
highest_rated = df.loc[highest_rated_idx]
print(f"评分最高的电影: {highest_rated['original_title']}")
print(f"评分: {highest_rated['vote_average']}")
print(f"投票人数: {highest_rated['vote_count']}")
```

```python
# 作业 2: 找出收入最高的电影
highest_revenue_idx = df['revenue'].idxmax()
highest_revenue = df.loc[highest_revenue_idx]
print(f"收入最高的电影: {highest_revenue['original_title']}")
print(f"收入: ${highest_revenue['revenue']:,}")
print(f"预算: ${highest_revenue['budget']:,}")
```

```python
# 作业 3: 数据覆盖的年份范围
earliest_year = df['release_year'].min()
latest_year = df['release_year'].max()
print(f"数据从 {earliest_year} 到 {latest_year}")
print(f"共 {latest_year - earliest_year + 1} 年")
```

```python
# 作业 4: 预算为 0 的电影数量
zero_budget = df[df['budget'] == 0]
zero_budget_count = len(zero_budget)
print(f"预算为 0 的电影有 {zero_budget_count} 部")
print(f"占总数的 {zero_budget_count / len(df) * 100:.1f}%")
```
