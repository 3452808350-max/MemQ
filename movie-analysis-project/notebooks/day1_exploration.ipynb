## 关键概念
- `pd.read_csv()`: 读取 CSV 文件
- `df.shape`: 数据形状（行数, 列数）
- `df.head()`: 查看前 N 行
- `df.columns`: 列名
- `df.dtypes`: 数据类型
- `df.describe()`: 统计摘要
- `df.isnull().sum()`: 缺失值统计
